O problema da mochila (em inglês, Knapsack problem) é situação em que é necessário preencher uma mochila com objetos de diferentes pesos e valores. O objetivo é que se preencha a mochila com o maior valor possível, não ultrapassando o peso máximo. Neste exemplo vamos considerar que só é possível colocar na mochila um objeto de cada tipo.

O primeiro desafio é pensar como representar uma mochila e os objetos em um programa de computador. Lembre-se que nesta etapa deve ser feita uma abstração, ou seja, somente as características relevantes para resolver o problema devem ser consideradas. 

```
class Objeto:
  def __init__(self, nome = 'default', peso = -1, valor = -1):
    self.nome = nome
    self.peso = peso
    self.valor = valor
  
  def setNome(self, nome):
    self.nome = nome
  
  def getNome(self):
    return self.nome

  def setPeso(self, peso):
    self.peso = peso
  
  def getPeso(self):
    return self.peso

  def setValor(self, valor):
    self.valor = valor

  def getValor(self):
    return self.valor
```

Para definir um objeto de forma computacional para o problema em questão só é necessário representar o seu nome, peso e valor. 

Assim sendo, foi utilizado o conceito de classe (Programação Orientada a Objetos) para descrever um objeto do problema. 

**LEMBRETE:** Uma **classe** pode ser definida como uma descrição das propriedades ou estados 
possíveis de um conjunto de objetos, bem como os comportamentos ou ações aplicáveis a estes mesmos objetos.
